```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lx = 1
f  = 10
nx = 1000
x = np.linspace(0, lx, nx)
y = np.sin(2*np.pi*f*x)
```

```python
plt.figure()
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
def finite_difference(f, x, eps, mode='central'):
    if mode is 'central':
        x1 = x-eps/2
        x2 = x+eps/2
    elif mode is 'forward':
        x1 = x
        x2 = x+eps        
    elif mode is 'backward':
        x1 = x-eps
        x2 = x
    else:
        raise ValueError('Differentiation mode not recognized!')
    
    dx = x2-x1
    df = f(x2)-f(x1)
    
    return df/dx
        
```
